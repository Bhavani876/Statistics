```python
#a
n=25
p=0.15
q=1-p
k=np.arange(0,5)
five_oreo_probabitity=st.binom.pmf(k,n,p)
```

```python
#method 1
sum(five_oreo_probabitity)
```

```python
# conclusion:-sum of fewer than five purchases.
```

```python
#method 2
five_oreo_probabitity=st.binom.cdf(4,n,p)
five_oreo_probabitity
```

```python
#b
n=25
p=0.15
q=1-p
k=np.arange(11,25)
ten_oreo_probabitity=st.binom.pmf(k,n,p)
```

```python
ten_oreo_probabitity
```

```python
# conclusion:- more than ten purchases.
```
