# 3.1.4 stratified sampling  for year

```python
year_grp=data.groupby(['Year'])
year_data=df['Year'].unique()
str_data=pd.DataFrame()
for i in year_data:
    year_d=year_grp.get_group(i)
    sam_year=year_d.sample(3)
    str_data=str_data.append(sam_year)
str_data
```

# 3.1.4 stratified sampling  for Genre

```python
Genre_grp=data.groupby(['Genre'])
Genre_data=data['Genre'].unique()
str_df=pd.DataFrame()
for i in Genre_data:
    Genre_d=Genre_grp.get_group(i)
    sam_Genre=Genre_d.sample(2)
    str_data=str_data.append(sam_Genre)
str_data
 
```

# 3.1.5 Clustered sampling

```python
year_grp=data.groupby(['Year'])
year_data=[2015,2018,2020]
clus_data=pd.DataFrame()
for i in year_data:
    year_d=year_grp.get_group(i)
    sam_year=year_d.sample(frac=.20)
    clus_data=clus_data.append(sam_year)
clus_data
```

```python
Genre_grp=data.groupby(['Genre'])
Genre_data=['Action','Puzzle']
clus_data=pd.DataFrame()
for i in Genre_data:
    Genre_d=Genre_grp.get_group(i)
    sam_Genre=Genre_d.sample(frac=.20)
    clus_data=clus_data.append(sam_Genre)
clus_data
```
